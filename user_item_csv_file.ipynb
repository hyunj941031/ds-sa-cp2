```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('/content/drive/MyDrive/fashion_campus_dataset/transaction_new.csv')
```

```python
data.tail()
```

```python
# 사용할 columns
columns = ['created_at', 'customer_id', 'product_id']

# 시간 관련 columns
df = data[columns]
df['datetime'] = pd.to_datetime(df['created_at'])
df['timestamp'] = df.datetime.values.astype(np.int64) // 10 ** 9

# 컬럼 drop
df = df.drop(columns = ['created_at', 'datetime'], axis=1)

# 정렬, 컬럼명 변경
df = df.sort_values(by=['customer_id','timestamp']).reset_index(drop=True).rename(columns={'customer_id' : 'userId', 'product_id' : 'itemId'})

df
```
